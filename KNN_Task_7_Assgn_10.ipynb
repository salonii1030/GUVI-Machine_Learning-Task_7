```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from scipy.spatial import distance
```

```python
train = pd.read_csv("trainKNN.txt", header=None)
train.columns = ['Id', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type']
train = train.drop('Id', axis=1)
test = pd.read_csv('testKNN.txt', header=None)
test.columns=['Id', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type']
test = test.drop('Id', axis=1)
```

```python
def standardize (df):
    for i in df.columns:
        if i != "Type": 
            df[i] = (df[i] - df[i].mean())/df[i].std()
    return df
```

```python
train = standardize(train)
test = standardize(test)
train.head()
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# train = scaler.fit_transform(train)
# test = scaler.transform(test)
```

```python
sqeucl_model = KNeighborsClassifier(n_neighbors = 4, metric=distance.sqeuclidean) 
manhattan_model = KNeighborsClassifier(n_neighbors = 4, metric=distance.cityblock) 
x_train = train.drop(["Type"], axis=1)
y_train = train["Type"]
sqeucl_model.fit(x_train,y_train)
manhattan_model.fit(x_train, y_train)
x_test = test.drop("Type", axis=1) 
y_test = test["Type"]
```

```python
man_pred = manhattan_model.predict(x_test)
```

```python
euc_pred = sqeucl_model.predict(x_test)
```

```python
df_pred = pd.DataFrame({"Actual": y_test, "Euclidean": euc_pred,"Manhattan":man_pred})
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print("Accuracy for Sq.Euclidean: ",accuracy_score(y_test, euc_pred))
print(classification_report(y_test, euc_pred))
print("\n")
print("Accuracy for Manhattan: ",accuracy_score(y_test,man_pred),"\n")
print(classification_report(y_test, man_pred))
```
